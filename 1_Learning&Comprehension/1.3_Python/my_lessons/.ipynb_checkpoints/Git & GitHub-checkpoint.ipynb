# What is Git? Why is it useful?

Git is a version control system. It maintains a history of all changes made to the code. The changes are stored in a special database called “repository”, also known as “repo”.
Git was also created by Linus Torvalds in the process of developing Linux.

- Tracking changes made to the software code.

- Allowing to work collaboratively.

Notes: 



- This history of the changes made to a project folder is tracked by the .git folder. So, the first thing you do is initialize the folder you want git to track.
To initialize a folder with git, use the command ```git init```

- Create a file. Say a text file using ```touch text.txt``` using the command line or use GUI if you prefer that. Once done, the question is if git is aware of the changes we made. Did it click a "hypothetical snapshot" of the current state
for us? 

- No. Currently that file and everything else is untracked. So if we are working on a collaborative space, nobody would know that text.txt exists and was added by me.

- Couple and guests wedding photo. A couple would call people whose pictures have not been taken to the stage, click a picture and save that picture in the history of the wedding album. The guests whose picture has not been taken in context of Git are called "untracked files".

- ```git add _filename_/.``` adds either the file name that you specify or all the files if you specify '.' to the staging area.

- Now, we can click their picture and store a snapshot in the git history. ```git commin -m "your message"```

- ```git status``` checks the current status of all the files in the folder.

- Imagine you deleted the file by mistake. You cannot delete a commit from the middle. So to roll back to say 3 commits ago, just copy the hash ID of the commit and use ```git reset hashcode```
Now we have unstaged the commit.

- Once the unstage is done, we can restore the unnecessary changes.

- For situations when there are changes that we do not want to let go and we do not want to commit also, we "backstage" them. ```git stash``` To bring them back, use ```git stash pop```. To permanently clear the "backstage" changes, use ```git stash clear```

# What is GitHub?

GitHub, Inc. is an Internet hosting service for software development and version control using Git. Git is the technology like "email tech" and GitHub is the platform like "Gmail."

Other such places are: BitBucket, GitLab.

Notes:

- ```git remote add origin (link)``` to connect local git repository to GitHub.

-  ```git remote -v``` displays all the URLs associated to your folder. 

-  ```git push origin main``` pushes all the changes to the "origin" URL and "main" branch from your local repo. The name of the main branch is main. Previously it was known as "master".

- Branches help you work on new features by taking a copy of the previous functioning code. 

- main branch contains code used by people (in open source). Therefore, the best practice is not to commit directly to the main branch. Also, making changes directly to the main branch could be dangerous for open source projects.

- When we have multiple branches, the HEAD is a pointer that controls on which branch will all the commits be added. 

- ```git merge (branch name)``` merges the feature branch to the main branch. Merging happens with a pull request. 

- You can create a copy of the project to your account using ```fork``` option on GitHub.

- The URL that we used to fork our projectis called upstream URL. Origin is our personal URL. 

- Pull request: 

Create a new branch -> Checkout(move HEAD to current branch) to the branch -> make changes 

How do we merge my branch to the main project main branch?

Any changes we make in the forked and branched version of the original project has to be requested to be merged with the original. Once you push your changes to your "branch name", your will see an option to merge an create a pull request on GitHub.

- to force push add a "-f" after push command
